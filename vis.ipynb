```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("train.csv")

data.at[1444, "consumption"] = (data.iloc[1443].consumption + data.iloc[1445].consumption) / 2
data.at[4972, "consumption"] = (data.iloc[4971].consumption + data.iloc[4973].consumption) / 2

for index, row in data.iterrows():
    data.at[index, "time"] = int(row.time[11] + row.time[12])

data.head(50)
```

```python
data = data.drop("prcp", axis=1)
data = data.drop("snow", axis=1)
data = data.drop("coco", axis=1)
data = data.drop("dwpt", axis=1)
data = data.drop("rhum", axis=1)
data = data.drop("wdir", axis=1)
data = data.drop("wspd", axis=1)
data = data.drop("wpgt", axis=1)
data = data.drop("pres", axis=1)
data = data.drop("el_price", axis=1)

X = data.loc[:, data.columns != "consumption"]
y = data.consumption

X.head()
```

```python
from time import time
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import ExtraTreesRegressor

gbr = GradientBoostingRegressor(loss="absolute_error", n_estimators=100)
etr = ExtraTreesRegressor(criterion="absolute_error", n_estimators=100)

gbr_time = np.zeros(8)
etr_time = np.zeros(8)

for i in range(8):
    start = time()
    gbr.fit(X.head((i+1)*1000), y.head((i+1)*1000))
    gbr_time[i] = time() - start
    print("gbr" + str(i))
    
for i in range(8):
    start = time()
    etr.fit(X.head((i+1)*1000), y.head((i+1)*1000))
    etr_time[i] = time() - start
    print("etr" + str(i))
```

```python
plt.figure(figsize=(15, 10))
plt.rcParams['font.size'] = 18
plt.xlabel("num of samples (thousands)")
plt.ylabel("training time (seconds)")
plt.plot(np.arange(1, 9), gbr_time, label="Gradient Boosting Regressor")
plt.plot(np.arange(1, 9), etr_time, label="Extra Trees Regressor")
plt.legend()
plt.savefig("train_time"+".png", facecolor=(1, 1, 1))
```
